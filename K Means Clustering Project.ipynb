```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

```python
finaldf = pd.get_dummies(df,columns=['Private'],drop_first=True)
```

```python
finaldf.head()
```

```python
finaldf.rename(columns={'Private_Yes':'Cluster'},inplace=True)
```

```python
finaldf.head()
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(finaldf['Cluster'],kmeans.labels_))
print('\n')
print(classification_report(finaldf['Cluster'],kmeans.labels_))
```
